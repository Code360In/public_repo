## Sparkクイックスタート(Python)
* Pythonを用いたデータフレーム操作をDatabricksノートブックで実行します。
* 参考 http://spark.apache.org/docs/latest/quick-start.html

```
# ファイルシステムを参照します
display(dbutils.fs.ls("/databricks-datasets/samples/docs/"))
```

データフレームには、新たなデータフレームへのポインタを返却する ***変換処理(transformations)*** と値を返却する ***アクション(actions)*** があります。

```
# 変換(transformation)
textFile = spark.read.text("/databricks-datasets/samples/docs/README.md")
```

```
# アクション(action)
textFile.count()
```

```
# テキストファイルの最初の行を出力
textFile.first()
```

ここでは、ファイルの一部を新たなデータフレームとして返却するフィルター ***変換処理(transformation)*** を用います

```
# データフレームの全ての行をフィルタリングします
linesWithSpark = textFile.filter(textFile.value.contains("Spark"))
```

上の処理がすぐに完了したことに注意してください。上は変換処理であり、アクションがまだ呼ばれていません。
* しかし、下のセルでアクション(例: count、take)を実行することで、処理が実行される様子を見ることになります。

```
# count(アクション)を実行します 
linesWithSpark.count()
```

```
# 最初の5行をアウトプットします
linesWithSpark.take(5)
```
